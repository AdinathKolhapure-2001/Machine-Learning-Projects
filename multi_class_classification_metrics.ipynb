1. Confusion Matrix

*   In Binary Classification

*   True Positive(TP) --> Predicted value is positive and the prediction is correct
*   True Negative(TN) --> Predicted value is negative and the prediction is correct
*   False Positive(FP) --> Predicted value is positive but the prediction is wrong
*   False Negative(FN) --> Predicted value is negative but the prediction is wrong


*   FP is known as type 1 error
*   FN is known as type 2 error


* From [medium blog](https://medium.com/analytics-vidhya/what-is-a-confusion-matrix-d1c0f8feda5) -->
![precision.webp](![accuracy.webp](![precision.webp](![confusion_mat_1.webp](accuracy.webp))))
* True Positives (TP): when the actual value is Positive and
  predicted is also Positive.
* True negatives (TN): when the actual value is Negative and prediction is also Negative.
* False positives (FP): When the actual is negative but prediction is Positive. Also known as the Type 1 error
* False negatives (FN): When the actual is Positive but the prediction is Negative. Also known as the Type 2 error

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
cm = confusion_matrix(y_test, log_model.predict(X_test), labels = log_model.classes_, normalize = None)
cm
```

```python
# Row wise normalization (true label wise normalization)
confusion_matrix(y_test, log_model.predict(X_test), labels = log_model.classes_, normalize = 'true')
```

```python
# Column wise normalization (prediction label wise normalization)
confusion_matrix(y_test, log_model.predict(X_test), labels = log_model.classes_, normalize = 'pred')
```

```python
# all normalization
confusion_matrix(y_test, log_model.predict(X_test), labels = log_model.classes_, normalize = 'all')
```

```python
# confusion matrix display
cm_disp = ConfusionMatrixDisplay(confusion_matrix = cm, display_labels = target_names)
cm_disp.plot()
```

## 2. Accuracy Score

*  Accuracy is the number of correct predictions from total predictions
*  We can calculate accuracy by comparing true values with predicted values
* We can also get the accuracy score from confusion matrix

![accuracy.jpg](image-39.png)

```python
# We can calculate accuracy by comparing true values with predicted values
y_pred = log_model.predict(X_test)
(y_test == y_pred).sum() / len(y_pred)
```

```python
# We can also get the accuracy score from confusion matrix
# from the above confusion matrix we can get the number of correct predictions
corr_pred = 14 + 7 + 8
total_pred = len(y_test)
corr_pred / total_pred
```

*  Accuracy is a valid choice of evaluation for classification problems which are well balanced and not skewed or there is no class imbalance.

*   When data is imbalanced accuracy score will be misleading
*   In that case we have to consider confusion matrix and other classification metrics



## 3. Precision

*  Precision is the proporation of correct prediction from the total positive prediction (Actual positives from predicted positives)
*  For multiclass classification problem we have to calculate the precision for each class
* precision = TP / (TP + FP)
* â€œPrecision is a useful metric in cases where False Positive is a higher concern than False Negativesâ€

precision.webp

```python
from sklearn.metrics import precision_score
```

```python
# This will return precision_score for each class
# If we want precision_score for specific class we can specify the list of class to the labels parameter
# precision_score(y_test, y_pred, labels = [0, 1, 2], average = None)
precision_score(y_test, y_pred, average = None)
```

### Micro Averaging
* Micro averaging computes a global average F1 score by counting the sums of the
* True Positives (TP), False Negatives (FN), and False Positives (FP)
* micro averaging is same as accuracy, it calculates the number of correctly classified datapoints


image-39.png

```python
precision_score(y_test, y_pred, average = 'micro')
```

### Macro-Averaging
* The macro-averaged is computed by taking the arithmetic mean (aka unweighted mean) of all the classes
* This method treats all classes equally regardless of their support values.

image-34.png


```python
precision_score(y_test, y_pred, average = 'macro')
```
