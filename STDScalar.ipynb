```python
import numpy as np
import pandas as pd
from sklearn import linear_model
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
scale = StandardScaler()
```

```python
df = pd.read_csv("/content/multi_lin_data_scale.txt", delimiter = ',')
df.head()
```

```python
df.info()
```

```python
df.corr(numeric_only = True).iloc[2:3, : ]
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[['Weight', 'Volume']], df['CO2'], test_size = 0.2)
```

```python
X_train_scaled = scale.fit_transform(pd.DataFrame(X_train))

y_train_scaled = scale.fit_transform(np.array(y_train).reshape(-1, 1))

X_test_scaled = scale.fit_transform(pd.DataFrame(X_test))

y_test_scaled = scale.fit_transform(np.array(y_test).reshape(-1, 1))
```

```python
reg = linear_model.LinearRegression()

reg_scaled = linear_model.LinearRegression()
```

```python

reg.fit(X_train, y_train)

reg_scaled.fit(X_train_scaled, y_train_scaled)
```

```python
y_pred = reg.predict(X_test)

y_pred_scaled = reg_scaled.predict(X_test_scaled)
```

```python
print("Actual: ", r2_score(y_test, y_pred))

print("Scaled: ", r2_score(y_test_scaled, y_pred_scaled))
```

```python
print("Actual: ", r2_score(y_test, y_pred))

print("Scaled: ", r2_score(y_test_scaled, y_pred_scaled))
```

```python
print("Actual: ", mean_squared_error(y_test, y_pred))

print("Scaled: ", mean_squared_error(y_test_scaled, y_pred_scaled))
```
