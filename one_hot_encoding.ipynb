```python
import numpy as np
import pandas as pd
```

```python
house_df = pd.read_csv('/content/homeprices.csv')
house_df
```

```python
# get_dummies works like one hot encoding
dummies = pd.get_dummies(house_df['town'])
dummies
```

```python
house_df_dummies = pd.concat([house_df, dummies], axis = 'columns')
house_df_dummies
```

```python
final_df = house_df_dummies.drop(columns=['town', 'west windsor'])
final_df
```

```python
X = final_df.drop(columns='price')
y = final_df['price']
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
X_train.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
y_pred
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_pred)
```

```python
# This also gives coefficient of determination(R^2) for linear regression
lr.score(X_test, y_test)
```

```python
car_df = pd.read_csv('/content/carprices.csv')
car_df.head()
```

```python
car_df.shape
```

```python
dummies = pd.get_dummies(car_df['Car Model'])
dummies
```

```python
merged = pd.concat([car_df, dummies], axis='columns')
merged
```

```python
final_df = merged.drop(columns = ['Car Model', 'Mercedez Benz C class'])
final_df
```

```python
X = final_df.drop(columns = 'Sell Price($)')
y = final_df['Sell Price($)']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 2)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
# 3) Tell me the score (accuracy) of your model. (Hint: use LinearRegression().score())
model.score(X_test, y_test)
```

```python
# 1) Predict price of a mercedez benz that is 4 yr old with mileage 45000
model.predict([[45000, 4, 0, 0]])
```

```python
# 2) Predict price of a BMW X5 that is 7 yr old with mileage 86000
model.predict([[86000, 7, 0, 1]])
```
