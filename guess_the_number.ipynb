```python
# Guess the number game

"""In this game computer will decide a random number between range. user have to
guess that number. if the guessed number is lower than the chosen number then
computer will promt that number is 'too low', if guessed number is greater than the chosen
number computer will prompt 'too high'. If the guesed number is correct then it will say
that you guessed the number right"""
```

```python
import math
import random
```

```python
def guess_the_number():

  # take the range from the user

  lower = int(input("Enter the lower range value: "))
  upper = int(input("Enter the upper range value: "))

  # Generate a radom number between lower and upper

  secret_number = random.randint(lower, upper)

  # Calculate the maximum number attempts

  max_attempts = math.ceil(math.log2(upper - lower + 1))

  attempts = 0
  guessed_num_list = []

  print("Welcome to guess the number game!")
  print(f"I have picked a number between {lower} and {upper}. Can you guess what it is?")
  print(f"You have maximum {max_attempts} attempts")

  while attempts < max_attempts:

    # Take the input from the user
    guess = int(input(f"Enter your guess (between {lower} and {upper}): "))


    # append the guessed number to the guessed_num_list
    guessed_num_list.append(guess)

    if guess < secret_number:
      print("Too low! Try again")
    elif guess > secret_number:
      print("Too high! Try again")
    else:
      print(f"Congratulations! you have guessed the number {secret_number}")
      break

    attempts += 1

    # Display the remainning number of attempts
    if max_attempts - attempts > 0:
      print(f"You have {max_attempts - attempts} remaining attempts")

  else:
    print(f"Sorry, you ran out of attempts! The secret number was {secret_number}")

  # print the guessed number list
  print(f"And here is the list of your guessed numbers: {guessed_num_list}")
```

```python
guess_the_number()
```
